## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set(style="darkgrid",font_scale=1.5)

from sklearn import datasets
from sklearn.cluster import KMeans
```

## Loading Dataset

```python
df = datasets.load_iris()
```

```python
df = pd.DataFrame(df.data, columns = df.feature_names)
```

```python
df.head()
```

---

## Findinging the optimum number of clusters for k-means classification

```python
x = df.iloc[:, [0, 1, 2, 3]].values

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

## Plotting the results onto a line graph, allowing us to observe 'The elbow'

```python
plt.figure(figsize=(15,5))
plt.plot(range(1, 11), wcss)
plt.title('The elbow method',fontweight="black",size=20,pad=15)
plt.xlabel('Number of clusters')
plt.xticks(list(range(0,11)))
plt.ylabel('WCSS') 
plt.show()
```

<div style="font-family:Verdana; background-color:aliceblue; padding:30px; font-size:17px;color:#034914">

<b>Observation</b><br>
    
ðŸ’¡ We can clearly see from the above graph that the optimum clusters is where the elbow occurs.<br>
ðŸ’¡ This is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration.<br>
ðŸ’¡ So can make an insight that the total number of clusters will be <b>3</b><br>
</div>

---

## Creating thee K-Means Clustering Model.

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

## Visualizing the Clusters Formed.

```python
plt.figure(figsize=(16,6))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 200, c = 'black', label = 'Centroids')
plt.legend()
plt.show()
```
